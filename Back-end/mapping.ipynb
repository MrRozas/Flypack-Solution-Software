```python
import folium

#https://www.youtube.com/watch?v=AeYxEDM1o2E&ab_channel=BugBytes

location1 = float(-33.4990878), float(-70.5520428)
location2 = float(-33.5079795), float(-70.5496573)
location3 = float(-33.4322916), float(-70.530344)

map = folium.Map(location=location1, width=1500, height=1000)

folium.Marker(location=location1, popup='Inicio entrega').add_to(map)
folium.Marker(location=location2, popup='Entrega 1').add_to(map)
folium.Marker(location=location3, popup='Entrega 2').add_to(map)
```
